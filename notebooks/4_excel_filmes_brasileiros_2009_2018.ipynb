## Análise de filmes exibidos no brasil de 2009 a 2018
com base nos dados da Ancine (https://oca.ancine.gov.br/cinema) Vamos analisar os filmes exibidos no Brasil

Arquivo -->filmes_exibidos_2009_2018.xlsx 

```python
# importando bibliotecas
import pandas as pd
# configurando para aparecer o gráfico na tela
%matplotlib inline
```

```python
# Criando o DataFrame
# Precisamos pular as duas primeiras linhas (skiprows)
# Precisamos descartar as 16 últimas linhas que contem informações inválidas usando (skipfooter)
df=pd.read_excel('filmes_exibidos_2009_2018.xlsx', skiprows=2, skipfooter=16)

# Exibe as 5 primeiras linhas do DataFrame
df.head()
```

```python
df.tail(20)
```

```python
# verificar todos os tipos das colunas 
df.dtypes
```

```python
# converter o campo Data de Lançamento em data
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Vamos tentar identificar o problema da coluna Data de lançamento, já que não conseguiu converter em data
df['Data de lançamento'].value_counts()
```

```python
# Exibir 5 filmes que estão com datas igual Relançamento
df[df['Data de lançamento'] == 'Relançamento'].head()
```

## Decisão sobre dados incorretos/faltantes/divergentes

```python
# Mostra o tamanho do DataFrame com formato de linhas e colunas
df.shape
```

```python
# Criar um novo DataFrame sem os relançamentos (!= significa diferente)
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
# tipos de colunas do df_novo
df_novo.dtypes
```

```python
# converter a coluna Data de lançamento para datetime
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
# mostrar tipos de coluna novamente
df_novo.dtypes
```

```python
# Quais os anos tiveram mais filmes lançados?
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Plotando um gráfico dos lançamentos por ano
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
# Qual filme tem maior bilheteria?
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# ver os 3 filmes com maior bilheteria
df_novo.nlargest(3,'Renda (R$) no ano de exibição')
```

```python
# Qual filme com menor bilheteria
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```

```python
# Quantos filmes brasileiros e estrangeiros
df_novo['Nacionalidade da obra'].value_counts()
```

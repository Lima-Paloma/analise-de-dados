# Análise dos dados de reembolso dos senadores - 2019

```python
# Importando bibliotecas
import pandas as pd
```

```python
# Carregar o arquivo csv com os dados dos senadores
df = pd.read_csv('2019_senadores_ceaps.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')

df.head()
```

```python
# Verificar tamanhos e tipos dos dados
df.info()
```

```python
# object é quando os dados são textos ou não reconhecidos
# Então mudo no pd.read, mudando para decimal=','
# Mostrar análise descritiva dos dados
df.describe()
```

```python
# Qual o total de reemblso
df['VALOR_REEMBOLSADO'].sum()
```

```python
#Quantos reembolsos foram solicitados por cada senador
df['SENADOR'].value_counts()
```

```python
# Agrupar todos os reembolsos para cada senador
# Quanto cada senador solicitou de reembolso em 2019?
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
# Quais são os 5 aiores valores de reembolso?
df.nlargest(5,'VALOR_REEMBOLSADO').T
```

```python
# Vamos ver os 5 menores reembolsos
df.nsmallest(5,'VALOR_REEMBOLSADO').T
```

```python
# Quantidade de reembolsos por tipo de despesa
df['TIPO_DESPESA'].value_counts()
```

```python
# Quanto foi reembolsado por cada tipo de despesa
# sort values ordena 
df.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

## Análise do Eleitorado Brasileiro 2020

Arquivo --> eleitorado_municipio_2020.csv

```python
# Importando Bibliotecas
import pandas as pd
```

```python
# Carregando o arquivo
df=pd.read_csv('eleitorado_municipio_2020.csv', sep = ';', encoding = 'latin1')

df.head(3).T
```

```python
# Verificar os tipos e quantidades dos dados
df.info()
```

```python
# Quais as cinco cidades com maior quantidade de eleitores deficientes
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitoras:', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores:', df['QTD_ELEITORES_MASCULINO'].sum())
print('Não Informado:', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Criar variáveis para calcular percentuais
total_eleitores = df['QTD_ELEITORES'].sum()
total_feminino = df['QTD_ELEITORES_FEMININO'].sum()
total_masculino = df['QTD_ELEITORES_MASCULINO'].sum()
total_nao_informado = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
# Mostrar os valores percentualmente % - arredondando para apenas 2 casas decimais
print('Eleitoras:', ((total_feminino/total_eleitores) * 100).round(2))
print('Eleitores:', ((total_masculino/total_eleitores) * 100).round(2))
print('Não Informado:', ((total_nao_informado/total_eleitores) *100).round(2))
```

```python
# Quantos municípios tem mais homens que mulheres
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Vamos criar uma coluna nova para indicar a relação feminino/masculino
# Basta fazer referência a uma coluna que não existe
df['RELACAO_FM_MASC'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# Quais são os municipios com maior relação feminino/masculino
df.nlargest(5,'RELACAO_FM_MASC').T
```

```python
# Quais municipios com a menor relação feminino/masculino
df.nsmallest(5,'RELACAO_FM_MASC').T
```

```python
# Criando um DataFrame somente com municípios brasileiros
# Usando o codigo do país
df_brasil = df[df['CD_PAIS'] == 1].copy()
df_brasil.head(3).T
```

```python
df_brasil.nlargest(5,'RELACAO_FM_MASC').T
```

```python
df_brasil.nsmallest(5,'RELACAO_FM_MASC').T
```

```python
# Para me mostrar o mapa
%matplotlib inline
# Plotar um gráfico da distribuição da relação feminino/masculino
# Bins significa alterar a quantidade de barras
df_brasil['RELACAO_FM_MASC'].plot.hist(bins=100)
```

```python
# Para fazer graficos em outra biblioteca seaborn - com graficos mais bonitos e mais faceis de alterar
# Importando biblioteca grafica
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotar a distribuição da relação feminino/masculino com seaborn
# Embelezando o gráfico, o kde tira a linha
sns.distplot(df_brasil['RELACAO_FM_MASC'], bins=100, color='red',kde=False)

# Colocando Título
plt.title('Relação Eleitoras/Eleitores', fontsize=18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.axvline(1.0, color='black', linestyle= '--')
```

```python
# Verificar a distribuição feminino/masculino, mostrando os pontos (muncípios)
# Dividindo a cada região do Brasil
sns.swarmplot(data=df_brasil, x='NM_REGIAO', y='RELACAO_FM_MASC')
# Traçando linha horizontal
plt.axhline(1.0, color='black', linestyle='--')
```

```python
# importando a bibliotecas 
# por convenção chamamos a biblioteca pandas de pd
import pandas as pd
```

```python
# carregar o arquivo csv
# pandas vai converter o arquivo em DataFrame 
# DataFrame é uma estrutura do pandas que se assemelha a uma planilha de excel
# por convenção chamamos o DataFrame de df
df = pd.read_csv('Unidades_da_Rede_Federal_de_EPCT.csv')
#função para visualizar a cabeça, as cinco primeiras linhas do dataframe
df.head()
```

```python
# quantas escolas no arquivo
df.count()
```

```python
# descreve apenas colunas númericas com estatísticas descritiva
df.describe()
```

```python
# mostra os tipos de cada coluna do dataframe, se texto, numérica e etc
df.dtypes
```

```python
# quantas escolas existem em cada estado UF
df['sigla_uf_unidade_ensino'].value_counts()
```

```python
# vamos contar quantas escolas tem em cada município
df['nome_municipio_unidade_ensino'].value_counts()
```

```python
# mostrando os dados de UF em um gráfico
# precisamos informar que queremos visualizar o gráfico dentro do próprio notebook
%matplotlib inline
# Agora vamos pegar as escolas por UF e mostrar em um gráfico de barras
df['sigla_uf_unidade_ensino'].value_counts().plot.bar()
```
